### Budget

```python
df = df[(df['Budget']!='Unknown')]
df
```

<details>
  <summary>Penjelasan :</summary>
Berdasarkan Pengecekan Inconsistent Data. Pada column <b>Budget</b> terdapat value <i>Unknown</i>, sehingga untuk menjaga kevalidan data dilakukan penghapusan data tersebut
</details>

```python
conversion_dict = {
    '$': 1.0,
    '€': 1.07,
    '£': 1.21,
    '₩': 0.00078,
    'C': 0.73,
    '₹': 0.012,
    '¥': 0.0075,
    'A': 0.68,
    'S': 0.095,
    'D': 0.14,
    'N': 0.10 
}
```

```python
for idx, value in enumerate(df['Budget']):
    if str(df.iloc[idx].Budget)[0] in conversion_dict.keys():
        df.loc[df['Budget'] == value, 'Budget'] = int(re.sub("[^0-9]", "", value)) * conversion_dict[df.iloc[idx].Budget[0]]
df['Budget'] = df['Budget'].astype(int)
```

```python
df.head()
```

<details>
  <summary>Penjelasan :</summary>
Berdasarkan Pengecekan Inconsistent Data. Pada column <b>Budget</b> memiliki value dengan keterangan mata uang, sehingga dilakukan konversi data agar data bisa dilakukan pengolahan
</details>
